```python
import seaborn as sns
import matplotlib as plt
import plotly as px
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
## scatter plot (usig 7 types of function together )-> it is a axis level function ->if we dont  want tp display in  Continuous means we will use scatter plot


sns.scatterplot(x='total_bill', y='tip', data=tips,hue="sex",style='time',size='size')
```

```python
## relplot (usig 7 types of function together ) -> figure level ->square shape -> if we dont  want tp display in  Continuous means we will use relpot

sns.relplot(x='total_bill', y='tip', kind = 'scatter', data=tips,hue='sex',style='time',size='size')
```

```python
## line plot -> basically it is scatr plot but -> for display  Continuous data  we will use "Line chart"

gap=px.data.gapminder()
gap
```

```python
sns.lineplot(x="year",y="lifeExp",data=gap[gap['country'] == 'India'])
```

```python
## using relpot
sns.relplot(x="year",y="lifeExp",kind='line',data=gap[gap['country'] == 'India'])
```

```python
## hue -> styles
temp_df=gap[gap['country'].isin(['India','Pakistan','china'])]
temp_df
```

```python
## using relplot
sns.relplot(x="year",y="lifeExp",kind='line',data=temp_df,hue='country')
```

```python
sns.lineplot(x="year",y="lifeExp",data=temp_df,hue='country')
```

```python
## for showing the continent

temp_df=gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
## for showing the continent

sns.relplot(x="year",y="lifeExp",kind='line',data=temp_df,hue='country',style='continent')
```

2.Distribution Plots
  
  plots under disyribution plot
    -> histplot
    -> kdeplot
    -> rugplot

```python
## histplot->histogram

## figure level -> is know as displot
## axes level -> histplot->kdeplot->rugplot
```

```python
## plotting histogram

tips
```

```python
sns.histplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'],kind='hist')
```

```python
## bins parameter
sns.displot(tips['total_bill'],kind='hist',bins=20)
```

```python
## its also possible to visualize the distributionof a categorical variable using the logic of a histogram
## Discrete bins are automatically set for categorical variables

tips
```

```python
# count plot based on the days

sns.displot(data=tips,x='day',kind='hist')
```

```python
## hue parameter
sns.displot(data=tips,x='tip',kind='hist',hue='sex')
```

```python
## element ->step
sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex')
```

```python
## faceting using col and row -> not work on histplot function
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex',col='sex')
```

```python
# kdeplot ->kernal density estimate plot
# Rather than using discrete bins, a kde plot smooths the observation with a gaussian kernal,producing a continuous density estimate
tips
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
## hue ->fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
## Rugplot

# Plot marginal distribution by drawing ticks along the x and y axes
# this function i s intended to complement other plots by showing the location of individual observation in an unobtrusive way

sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# a bivariate histogram  bins the data within rectangles that tile the plot
# and then shows the count of observation within each rectangle with the fill color

sns.histplot(data=tips,x='total_bill',y='tip')
##sns.displot(data=tips,x='total_bill',y='tip',kind='hist')
```

```python
# Bivaraiate Kdeplot
# a bivariate kde plot smoothes the (x,y) observation with ina 2d gaussin


sns.kdeplot(data=tips,x='total_bill',y='tip')
```

Matrix pol

  1-Heatmap
  2-Clustermap



```python
# Heatmap
gap
```

```python

# plot rectunglar  data as a color-encoded matrix
temp_df=gap.pivot(index="country",columns='year',values='lifeExp')

# axes level function
plt.pyplot.figure(figsize=(10,10))
sns.heatmap(temp_df)
```

```python
# annot
temp_df=gap[gap['continent']=='Europe'].pivot(index="country",columns='year',values='lifeExp')

plt.pyplot.figure(figsize=(10,10))
sns.heatmap(temp_df,annot=True,linewidths=0.5,cmap='viridis')
```

```python
# CLustermap

# plot a matrix dataset as a hierarchically-clustered heatmap
# this function requies scipy to be available.

iris=px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
